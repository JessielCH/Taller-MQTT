```python
import psutil
import json
import time
import threading
import paho.mqtt.client as mqtt

def conectar_mqtt(host, port, topic, mqtt_client):
    mqtt_client.connect(host, port, 60)
    mqtt_client.loop_start()

def enviar_datos_maquina2(mqtt_client, mqtt_topic):
    while True:
        cpu_percent = psutil.cpu_percent(interval=1)
        ram_percent = psutil.virtual_memory().percent
        disk_percent = psutil.disk_usage('/').percent
        net_stats = psutil.net_io_counters()

        datos_maquina2 = {
            "id_maquina": "maquina2",
            "tipo": "rendimiento",
            "datos": {
                "CPU": cpu_percent,
                "Memoria": ram_percent,
                "Disco": disk_percent,
                "Recepcion": {
                    "bytes_enviados": net_stats.bytes_sent,
                    "bytes_recibidos": net_stats.bytes_recv
                }
            }
        }

        # Mensaje de impresión antes de enviar los datos
        print("Enviando datos desde la Máquina 2:", datos_maquina2)

        # Enviar los datos de la Máquina 2 a la Máquina 1
        mqtt_client.publish(mqtt_topic, json.dumps(datos_maquina2))
        time.sleep(1)

mqtt_host = "broker.hivemq.com"
mqtt_port = 1883
mqtt_topic = "probar_1"

mqtt_client_maquina2 = mqtt.Client()
conectar_mqtt(mqtt_host, mqtt_port, mqtt_topic, mqtt_client_maquina2)

# Iniciar hilo para enviar datos desde la Máquina 2 a la Máquina 1
enviar_datos_maquina2_thread = threading.Thread(target=enviar_datos_maquina2, args=(mqtt_client_maquina2, mqtt_topic))
enviar_datos_maquina2_thread.start()

# Mantener el script principal en ejecución
enviar_datos_maquina2_thread.join()
```
