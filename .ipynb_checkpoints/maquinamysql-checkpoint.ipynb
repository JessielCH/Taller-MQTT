```python
import json
import platform
import time
import threading
import paho.mqtt.client as mqtt
import psutil
import mysql.connector
from mysql.connector import errorcode

# Función para conectar al servidor MQTT
def conectar_mqtt(host, port, topic, mqtt_client):
    mqtt_client.connect(host, port, 60)
    mqtt_client.loop_start()

# Función para obtener información del sistema
def obtener_info_sistema():
    try:
        system_info = {
            "Sistema": platform.system(),
            "Release": platform.release(),
            "Procesador": platform.processor(),
            "Nombre Sistema Operativo": platform.system_alias(platform.system(), platform.release(), 0),
            "Arquitectura": platform.architecture()[0]
        }
        return system_info
    except Exception as e:
        print(f"Error al obtener información del sistema: {e}")
        return None

# Función para enviar datos a través de MQTT y almacenarlos en MySQL
def enviar_datos_maquina1(mqtt_client, mqtt_topic, mysql_cursor, machine_id):
    while True:
        try:
            cpu_percent = psutil.cpu_percent(interval=1)
            ram_percent = psutil.virtual_memory().percent
            rom_percent = psutil.disk_usage('/').percent
            info_sistema = obtener_info_sistema()
            fecha = time.strftime('%Y-%m-%d %H:%M:%S')

            # Datos para MQTT
            datos_maquina1_mqtt = {
                "MachineID": machine_id,
                "datos": {
                    "CPU": cpu_percent,
                    "RAM": ram_percent,
                    "ROM": rom_percent,
                    "Info_Sistema": info_sistema,
                    "fecha": fecha
                }
            }

            # Enviar datos de la Máquina 1 a través de MQTT
            mqtt_client.publish(mqtt_topic, json.dumps(datos_maquina1_mqtt))

            # Datos para MySQL
            insert_query = "INSERT INTO `rendimientopc` (`MachineID`, `CPU`, `RAM`, `ROM`, `SYSTEM`, `FECHA`) " \
                           "VALUES (%s, %s, %s, %s, %s, %s)"
            data_mysql = (machine_id, cpu_percent, ram_percent, rom_percent, str(info_sistema["Nombre Sistema Operativo"]), fecha)

            # Ejecutar la sentencia SQL para insertar en MySQL
            mysql_cursor.execute(insert_query, data_mysql)
            mysql_connection.commit()

            time.sleep(1)

        except Exception as e:
            print(f"Error en enviar_datos_maquina1: {e}")

# Configuración MQTT
mqtt_host = "broker.hivemq.com"
mqtt_port = 1883
mqtt_topic = "grupo_7"
mqtt_client_maquina1 = mqtt.Client()

# Configuración MySQL
mysql_host = "mysql-arquitectura1.alwaysdata.net"
mysql_user = "349016_jj"
mysql_password = "0987074675"
mysql_database = "arquitectura1_uce"

try:
    mysql_connection = mysql.connector.connect(
        host=mysql_host,
        user=mysql_user,
        password=mysql_password,
        database=mysql_database
    )
    mysql_cursor = mysql_connection.cursor()
except mysql.connector.Error as err:
    print(f"Error en la conexión a MySQL: {err}")
    exit(1)

# Especifica el ID de la máquina
machine_id = "maquina3"

# Iniciar hilo para enviar datos desde la Máquina 1 a través de MQTT y MySQL
enviar_datos_maquina1_thread = threading.Thread(target=enviar_datos_maquina1, args=(mqtt_client_maquina1, mqtt_topic, mysql_cursor, machine_id))
enviar_datos_maquina1_thread.start()

# Inicialización de la conexión MQTT
conectar_mqtt(mqtt_host, mqtt_port, mqtt_topic, mqtt_client_maquina1)

# Mantener el script principal en ejecución
enviar_datos_maquina1_thread.join()
```
