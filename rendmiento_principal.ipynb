```python
import json
import psutil
import time
import paho.mqtt.client as mqtt

def conectar_mqtt(host, port, topic, mqtt_client):
    mqtt_client.connect(host, port, 60)
    mqtt_client.subscribe(topic)
    mqtt_client.loop_start()

def on_message(client, userdata, msg):
    response = json.loads(msg.payload)
    id_maquina = response["id_maquina"]
    tipo = response["tipo"]
    datos = response["datos"]

    print(f"Datos recibidos de {id_maquina} - Tipo: {tipo}")
    print(f"Uso de CPU: {datos['CPU']}%, Uso de RAM: {datos['Memoria']}%, Uso de Disco: {datos['Disco']}%")
    print(f"Bytes enviados: {datos['Recepcion']['bytes_enviados']}, Bytes recibidos: {datos['Recepcion']['bytes_recibidos']}")

mqtt_client_monitor = mqtt.Client()
mqtt_host = "broker.hivemq.com"
mqtt_port = 1883
mqtt_topic = "probar_1"

conectar_mqtt(mqtt_host, mqtt_port, mqtt_topic, mqtt_client_monitor)
mqtt_client_monitor.on_message = on_message

while True:
    # Obtener y mostrar datos locales de la Máquina 1
    cpu_percent = psutil.cpu_percent(interval=1)
    ram_percent = psutil.virtual_memory().percent
    disk_percent = psutil.disk_usage('/').percent
    net_stats = psutil.net_io_counters()

    print("Datos locales de la Máquina 1:")
    print(f"Uso de CPU: {cpu_percent}%, Uso de RAM: {ram_percent}%, Uso de Disco: {disk_percent}%")
    print(f"Bytes enviados: {net_stats.bytes_sent}, Bytes recibidos: {net_stats.bytes_recv}")

    time.sleep(1)  # Mantén el script en ejecución para recibir mensajes continuos
```
